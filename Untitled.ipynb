```python
x = df.drop('Exited',axis = 1)
y = df['Exited']
```

Although the accuracy is high, as the dataset is imbalanced, we can't solely rely on accuracy score as a metric for evaluation. So we tested the precision score, recall and f1 score as well, which resulted in drastic fall compared to accuracy. This required immediate attention that we should handle our imbalanced dataset

### Handling imbalanced dataset using SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
x_resampled,y_resampled = SMOTE().fit_resample(x,y)
```

```python
x_resampled.value_counts()
```

### Splitting Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42,stratify = y)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,precision_score,recall_score,f1_score
```

```python
log = LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
y_pred_log = log.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_log)
```

```python
print(classification_report(y_test,y_pred_log))
```

```python
precision_score(y_test,y_pred_log)
```

```python
recall_score(y_test,y_pred_log)
```

```python
f1_score(y_test,y_pred_log)
```

### SVC

```python
svm.fit(x_train,y_train)
```

```python
y_pred_svm = svm.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_svm)
```

```python
precision_score(y_test,y_pred_svm)
```
